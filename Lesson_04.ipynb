```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def func(a,x):
    return np.sin(a*x)
x = np.linspace(100,500,1001)
plt.plot(x,func(0.01,x), color="orange")
plt.plot(x,func(0.02,x), color="green")
plt.plot([100,500],[0,0], color='red', linewidth = 0.75, linestyle='--')
```

```python
def func(k,a):
    return np.pi*k/a
a = np.linspace(0.01,0.02,1001)
plt.plot(a,func(1,a), color="orange")
plt.plot(a,func(2,a), color="green")
plt.plot(a,func(3,a), color="blue")
#plt.plot([0.01,0.02],[500,500], color='red', linewidth = 0.75, linestyle='--')
plt.ylim([100,500])
plt.xlabel("a")
plt.ylabel("x")
plt.grid(True)
plt.show()
# Решением уравнения будут 3 графика ниже в указанных областях

# Несколько решений в численном виде:
alpha = 0.014
x = func(1,alpha)
x1 = func(2,alpha)
print (x, x1)
# проверка решения
y = np.sin(alpha*x)
y1=np.sin(alpha*x1)
print (y,y1)
```
